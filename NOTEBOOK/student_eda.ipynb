## EXPLORATORY DATA ANALYSIS AND FEATURE ENGINEERING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('../DATA/student-mat-cleaned.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
## PLOTTING A CORRELATION MATRIX
plt.figure(figsize=(10, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
```

```python
## CATEGORICAL FEATURES VS G3

categorical_features = ['sex','schoolsup','famsup','paid','activities','internet']
for feature in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=feature, y='G3', data=df)
    plt.title(f'G3 vs {feature}')
    plt.show()
```

```python
# NUMERICAL FEATURES VS G3
numerical_features =['age','traveltime','studytime','failures','absences','freetime','goout','Dalc','Walc','health','G1','G2']

for feature in numerical_features:
    plt.figure(figsize=(10, 5))
    sns.scatterplot(x=feature, y='G3', data=df)
    plt.title(f'G3 vs {feature}')
    plt.show()
```
